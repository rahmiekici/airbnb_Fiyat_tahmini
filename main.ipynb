```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
data = pd.read_csv("AB_NYC_2019.csv")
```

```python
data.columns
```

```python
data.head()
```

```python
print('Sutun Sayısı: {}'.format(data.shape[1]))
print('Satır Sayısı: {}'.format(data.shape[0]))
```

```python
data.isnull().sum()
```

```python
#data.drop(['name','id', 'host_id','last_review','host_name','neighbourhood','number_of_reviews','reviews_per_month','availability_365','calculated_host_listings_count'], axis = 1, inplace = True)
data.drop(['name','id', 'host_id','last_review','host_name','number_of_reviews','reviews_per_month','availability_365','calculated_host_listings_count'], axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
# Her sütun için ayrı encoder
le_neighbourhood = LabelEncoder()
le_room_type = LabelEncoder()
le_neighbourhood_group = LabelEncoder()
# Encode işlemleri
data['neighbourhood_group'] = le_neighbourhood_group.fit_transform(data['neighbourhood_group'])
data['neighbourhood'] = le_neighbourhood.fit_transform(data['neighbourhood'])
data['room_type'] = le_room_type.fit_transform(data['room_type'])
```

```python
data.head(10)
```

```python
data.info()
```

```python

Q1 = data.quantile(0.25, numeric_only=True)
Q3 = data.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

numeric_columns = data.select_dtypes(include=[np.number]).columns
data = data[~((data[numeric_columns] < (Q1 - 1.5 * IQR)) | (data[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
data.info()
```

```python
#Regression analysis to predict the price
#drop unneceassy columns
X = data.iloc[:,[0,1,2,3,4,6]]
y = data['price']
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.2, random_state= 0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Prepare a Linear Regression Model

reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
#R2 score
from sklearn.metrics import r2_score, mean_squared_error
r2=r2_score(y_test,y_pred)*100
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print("R2 score: ",r2)
print("RMSE: ",rmse)

#Error
error_diff = pd.DataFrame({'Gerçek Değerler': np.array(y_test).flatten(), 'Tahmin Edilen Değer': y_pred.flatten()})
print(error_diff.head(5))

#Visualize the error
df1 = error_diff.head(25)
df1.plot(kind='bar',figsize=(10,7))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
#ornek
ornek_veri = pd.DataFrame({
    'neighbourhood_group': [1],
    'neighbourhood': [108],
    'latitude': [40.75362],
    'longitude': [-73.98377],
    'room_type': [1],
    'minimum_nights': [3]
})

ornek_veri_scaled = scaler.transform(ornek_veri)

tahmin = reg.predict(ornek_veri_scaled)
print("Tahmini fiyat:", tahmin[0],"$")
```

```python
results_df = pd.DataFrame(columns=["Model", "R2 Score (%)", "RMSE","Tahmin Fiyatı"])
```

```python
results_df = pd.concat([results_df, pd.DataFrame([{
    "Model": 'Linear Regression Model',
    "R2 Score (%)": r2,
    "RMSE": rmse,
    "Tahmin Fiyatı" : tahmin[0]
}])], ignore_index=True)
```
